Task A : Lambda + .apply()

Instructions (must use lambda + .apply()):

Replace missing values in math, english, and science with the mean of each column

Create a new column called average that contains the average score of each student

Create a new column called passed:

True if average â‰¥ 70

False otherwise

Rules:

Use .apply() with lambda

Do not write custom functions

Do not use loops

```python
import pandas as pd

df = pd.DataFrame({
    "student": ["Ali", "Sara", "Omar", "Lina", "Huda", "Tariq"],
    "math": [90, 78, 65, 88, None, 55],
    "english": [85, 80, None, 92, 75, 60],
    "science": [88, None, 70, 90, 80, 58]
})
```

```python
df[['math', 'english', 'science']] = df[['math', 'english', 'science']].apply(lambda col: col.fillna(col.mean()))

df['average'] = df[['math', 'english', 'science']].apply(lambda row: row.mean(), axis=1)

df['passed'] = df['average'].apply(lambda x: x >= 70)

print(df)
```

Task B : Using Functions

Instructions (must use regular functions):

Write a function called fill_missing(column)

- Takes a pandas column

- Replaces missing values with the column mean

- Returns the cleaned column

Write a function called calculate_average(row)

- Takes a row

- Returns the average of math, english, and science

Use these functions to:

- Clean the DataFrame

- Add a column called average

Rules:

Use def to define functions

Use .apply() to apply your functions

No lambda expressions in this task

```python
def fill_missing(column):
    return column.fillna(column.mean())

def calculate_average(row):
    return row[['math', 'english', 'science']].mean()

df[['math', 'english', 'science']] = df[['math', 'english', 'science']].apply(fill_missing)

df['average'] = df.apply(calculate_average, axis=1)

print(df)
```
